- 한국의 경우 통합대기환경지수(CAI) 계산 방식은 6가지의 대기오염물질 지수 점수 중 가장 높은 점수를 통합지수값으로 사용함
- 초미세먼지(PM2.5), 미세먼지(PM10), 오존(O3), 이산화질소(NO2), 일산화탄소(CO), 아황산가스(SO2) 등 대표적인 6가지 오염물질에 대한 대기질의 상태지수를 제공한다.

# 6가지 오염물질
1. 미세먼지(PM10)<br>
우리 눈에 보이지 않는 아주 작은 물질로 대기 중에 오랫동안 떠다니거나 흩날려 내려오는 직경 10㎛ 이하의 입자상 물질을 말한다. 석탄, 석유 등의 화석연료가 연소될 때 또는 제조업·자동차 매연 등의 배출가스에서 발생한다. 이는 기관지를 거쳐 폐에 흡착되어 각종 폐질환을 유발하는 대기오염물질이다.

2. 초미세먼지(PM2.5)<br>
지름 2.5μm 이하의 초미세먼지 입자로 황산염·질산염·암모니아 등의 이온 성분과 금속화합물, 탄소화합물 등의 유해물질로 이루어져 있다. 주로 자동차 배기가스 등에서 발생하는 초미세먼지는 호흡기 깊숙이 침투해 폐 조직에 붙어 호흡기 질환을 일으키며, 혈관으로 흡수돼 뇌졸중이나 심장질환의 원인이 되는 물질이다. 

3. 오존(O₃)<br>
질소산화물과 탄화수소가 자외선에 의한 촉매반응을 하여 생성되는 물질로 2차 오염물질에 속한다. 오존농도가 0.1PPM을 넘어서면 불쾌감을 주는 것은 물론 폐기능 약화, 기침 증가, 눈 따가움 증상 등이 나타난다. 또한 식물 수확량 감소에도 영향을 미친다.

4. 이산화질소(NO₂)<br>
대기 중에서 일산화질소의 산화에 의해서 발생하는 것으로, 고농도에 노출되면 만성 기관지염, 폐렴, 폐출혈 등이 발생할 수 있다.

5. 일산화탄소(CO)<br>
무색·무취의 기체로서 산소가 부족한 상태에서 석탄이나 석유 등 연료가 탈 때 발생한다. 체내로 들어온 일산화탄소는 산소 대신 헤모글로빈과 결합하고, 산소가 체내에 공급되지 못하면서 저산소증을 유발시키는데, 이를 ‘일산화탄소 중독'이라 한다.

6. 아황산가스(SO₂)<br>
황이 연소할 때에 발생하는 기체로, 황과 산소의 화합물이다. 자극성 있는 냄새가 나는 무색 기체이고 인체의 점막을 침해하는 독성이 있다. 고농도에 노출되면 콧물, 땀, 기침이 나며 목구멍이나 가슴이 아프고, 호흡이 곤란해진다. 또 기관지염, 폐수종, 폐렴 등의 발생 가능성도 높인다.

# 통합대기환경지수 산출
- 통합대기환경지수 산출은 각 오염물질별로 점수를 산정하는데, 0∼500까지의 지수를 4단계(0~50, 51~100, 101~250, 251~)로 나눈다. 즉, 대기의 상태를 0에서 500점까지 수치화하는데 점수가 높아질수록 대기 상태가 좋지 않음을 의미한다. 지수 계산 방법은 6가지의 대기오염물질 지수 점수 중 가장 높은 점수를 통합지수값으로 사용한다.

![이미지](./통합환경지수(지수산출에_필요한_변수).jpg)

- 통합대기환경지수의 표현방식은 지수 점수 외에 색상이나 지수 구간별 픽토그램(pictogram)을 사용하여 일반인이 이해하기 쉽도록 제공되고 있다. 예컨대 ‘좋음’은 파랑, ‘보통’은 초록, ‘나쁨’은 노랑, ‘매우 나쁨’은 빨강으로 표시한다.

![이미지](./픽토그램.jpg)

- 여기서 '좋음', '보통', '나쁨', '매우 나쁨'은 다음과 같은 뜻을 담고 있다. 

◎ 좋음: 지수구간 A로, 대기 오염 관련 질환자군에서도 영향이 유발되지 않을 수준 
◎ 보통: 지수구간 B로, 환자군에게 만성 노출 시 경미한 영향이 유발될 수 있는 수준 
◎ 나쁨: 지수구간 C로, 환자군 및 민감군(어린이, 노약자 등)에게 유해한 영향을 유발하고, 일반인도 건강상 불쾌감을 경험할 수 있는 수준 
◎ 매우 나쁨: 지수구간 D는 환자군 및 민감군에게 급성 노출 시 심각한 영향을 유발하고, 일반인도 약한 영향이 유발될 수 있는 수준. 지수구간 E는 환자군 및 민감군에게 응급조치가 발생되거나, 일반인에게 유해한 영향이 유발될 수 있는 수준

## 학습데이터와 테스트데이터 분리하기

```python
from math import floor
```

```python
def train_test_split_df(df, test_size):
    # TODO: test_size(0 < test_size < 1) 비율에 따라 테스트/학습 데이터프레임으로 나누어 반환
    div = floor(df.shape[0] * (1 - test_size))
    return df.loc[:div], df.loc[div + 1:]
```

```python
train_df, test_df = train_test_split_df(df, 0.3)
```

## Prophet 모델 생성 및 학습 데이터 학습

```python
from fbprophet import Prophet
```

```python
model = Prophet(changepoint_range=1.0)
model.fit(train_df)
```
